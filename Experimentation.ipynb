```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

from KerasModel import Model
```

```python
train = pd.read_csv('Dataset/train.csv')
validation = pd.read_csv('Dataset/validation.csv')

X_train, y_train = train.drop('label',axis=1), train['label']
X_val, y_val = validation.drop('label',axis=1), validation['label']
```

```python
def experiment(n):
    layers = [n,n,n,n,n]
    nodes = [100,150,200,250,300]

    history_dict = {
        'model_name': [],
        'val_loss': [],
        'val_acc': []
    }


    for layer, node in zip(layers,nodes):
        model = Model(X_train, y_train)
        model.build_model(layers=layer, nodes=node)

        model.train_model(X_val, y_val, patience=3, epochs=30)

        history_dict['model_name'].append(model.model_name)
        history_dict['val_loss'].append(model.val_loss_history)
        history_dict['val_acc'].append(model.val_acc_history)
        
    return history_dict

def visualize(history):        
    history_df = pd.DataFrame(history)
    plt.figure(figsize=(16,6))
    for i in range(5):
        plt.plot(history_df['val_loss'][i])

    plt.xlabel('Epochs')
    plt.ylabel('Validation Loss')

    plt.legend(history_df['model_name'])
    plt.show()
```

```python
experiment_with_2layers = experiment(2)
```

```python
experiment_with_3layers = experiment(3)
```

```python
experiment_with_4layers = experiment(4)
```

```python
visualize(experiment_with_2layers)
```

```python
visualize(experiment_with_3layers)
```

```python
visualize(experiment_with_4layers)
```
